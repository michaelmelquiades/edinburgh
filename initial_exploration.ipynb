```python
ref = pd.read_csv('GS_Oct23.csv')
ref
```

# Try to make a feature/target dataset

# Try without any treatment of the data first

```python
data = dict(sample = [], 
            mean = [], std = [],
            maxi = [], mini = [], 
            wave_maxi = [], wave_mini = [])
for i in ref.id:
    s = str(i)
    data['sample'] += [s]
    data['mean'] += [df[s].mean()]
    data['std'] += [df[s].std()]
    data['maxi'] += [df[s].max()]
    data['mini'] += [df[s].min()]
    data['wave_maxi'] += [df[df[s] == df[s].max()].wavelength.values[0]]
    data['wave_mini'] += [df[df[s] == df[s].min()].wavelength.values[0]]
```

```python
df_poss = pd.DataFrame(data)
df_poss['cp'] = ref.CP
df_poss['ndf'] = ref.NDF
df_poss
```

# Try some simple regression techniques

```python
from sklearn.linear_model import LinearRegression
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import train_test_split, RepeatedKFold, cross_val_score
```

```python
x = df_poss.iloc[:, 1:5]
y = df_poss.iloc[:, -2]#[:, 7:]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
lin = LinearRegression()#MultiOutputRegressor(LinearRegression())
```

```python
cv = RepeatedKFold(n_splits = 10, n_repeats = 5, random_state=42)
scores = cross_val_score(lin, x, y, scoring='neg_mean_absolute_error', cv = cv, n_jobs = -1)
scores = np.absolute(scores)
print(f'MAE: {np.mean(scores)} ({np.std(scores)})')
```

# Stuff from lit reviews

```python
from sklearn.preprocessing import scale 
from sklearn.cross_decomposition import PLSRegression
from sklearn.metrics import mean_squared_error
```

```python
mae = []
for i in range(1, 5, 1):
    pls = PLSRegression(n_components = i)
    scores = cross_val_score(pls, x, y, scoring='neg_mean_absolute_error', cv = cv, n_jobs = -1)
    scores = np.absolute(scores)
    mae.append(scores.mean())
```

```python
plt.plot(mae)
plt.xticks(range(1, 5, 1))
plt.show()
```
